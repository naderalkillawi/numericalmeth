```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Constants
G = 6.67430e-11  # gravitational constant, m^3 kg^−1 s^−2
mass_sun = 1.989e30  # mass of the sun, kg
mass_earth = 5.972e24  # mass of the Earth, kg
AU = 149.6e9  # astronomical unit, meters

# Initial conditions
initial_position = np.array([AU, 0.0])  # initial position in x-y plane
initial_velocity = np.array([0.0, 29783.0])  # initial velocity in x-y plane

# Function to compute derivatives for the differential equations
def planetary_motion(t, state):
    position = state[:2]
    velocity = state[2:]
    r = np.linalg.norm(position)
    acceleration = -G * mass_sun * position / r**3
    return np.concatenate([velocity, acceleration])

# Time span
t_span = (0, 3.15e7)  # 1 year in seconds

# Solve the differential equations
solution = solve_ivp(planetary_motion, t_span, np.concatenate([initial_position, initial_velocity]), method='RK45', dense_output=True)

# Extracting results
time = solution.t
positions = solution.y[:2, :]
velocities = solution.y[2:, :]

# Plotting the orbit
plt.figure(figsize=(8, 8))
plt.plot(positions[0, :] / AU, positions[1, :] / AU)
plt.scatter(0, 0, color='yellow', marker='o', label='Sun')
plt.scatter(positions[0, 0] / AU, positions[1, 0] / AU, color='blue', marker='o', label='Initial Position')
plt.scatter(positions[0, -1] / AU, positions[1, -1] / AU, color='red', marker='o', label='Final Position')
plt.title('Orbit Simulation')
plt.xlabel('X (AU)')
plt.ylabel('Y (AU)')
plt.legend()
plt.grid(True)
plt.show()
```
